1) [$k$-nearest neighbor regression, 25 points] In $k$-NN regression the predicted value for point $\hat{x}$ is obtained by averaging over the $y$-values that correspond to the $k$ $x$-values that are closest to $\hat{x}$ using $d(a,b)=||a-b||$ as a measure distance between points $a$ and $b$.

Create a function which implements $k$-NN regression using NumPy functionality. Apply your function to the global warming data in `global.csv` (see below) to predict temperature fluctuations on a monthly basis between 1850 and 2015, using $K=2$. Create a scatter plot of the original data points and overlay with the predicted values (using a different color).

*Comments*: The global warming data set contains global annual temperature measurements from 1850 to 2015 (on the Celcius scale). The values have been scaled so that they represent temperature anomalies relative to the 1961 - 1990 average. The data comes from https://cdiac.ess-dive.lbl.gov/trends/temp/jonescru/jones.html

The following functions are useful: `numpy.linalg.norm`, `numpy.mean` and `numpy.argsort`.

```python
import numpy as np
import matplotlib.pyplot as plt
def knn_reg_predict(X, y, xpred, K):
    # k-nearest neighbor regression
    #    X ....... Input variables
    #    y ....... Output variable
    #    xpred ... Point to predict
    #    K ....... Number of neighbors (positive integer)
    # Return value
    #    ypre .... Predicted value at xpred
    assert(K>0 and int(K)==K)
    #index = X.index(xpred)
    n = len(X)
    xpred = (xpred-min(X))/(max(X)-min(X)) #Skölum xpred á [0,1] líkt og X
    X = (X-min(X))/(max(X)-min(X)) #Skölum X gögnin á [0,1]
    dist = np.zeros(n)
    
    for i in range(0,n):
        dist[i] =np.linalg.norm(X[i]-xpred)
    ysort = y[np.argsort(dist)]
    ypred = np.mean(ysort[:K])
    return ypred

X,y = np.loadtxt('global.csv', delimiter=',', unpack=True,skiprows=1)
K=2
plt.scatter(X,y)
xpred = np.arange(int(min(X)),int(max(X)),1/12)
m = len(xpred)
ypred = np.zeros(m)
for i in range(0,m):
    ypred[i] = knn_reg_predict(X,y,xpred[i],K)
plt.plot(xpred,ypred,c='g')
```

2) [Locally weighted regression, LWR, 25 points] In LWR the predicted value for point $\hat{x}$ is obtained by fitting a local regression model to the data. This is done by assigning individual weights to points in the data set, depending on how close they are to $\hat{x}$ and minimizing the following cost function
$$
J(\theta) = \frac{1}{2}\sum_{i=1}^n w_i (\theta^T x^{(i)} - y^{(i)})^2
$$
where $w_i > 0$ is the weight given to point $x^{(i)}$. We will use
$$
   w_i = \exp\left({-\frac{||\hat{x} - x^{(i)}||^2_2}{2\sigma^2}}\right), \qquad i=1,\ldots n
$$
where $\sigma>0$ is a parameter that needs to be specified.

Note that the values of the $w_i$'s depend on the point $\hat{x}$ we are predicting. It can be shown that the paramaters $\theta$ that minimize the cost function can be obtained by solving a system of linear equations
$$
(X^T W X)\theta = (X^T W)y
$$
where $W$ is a $n \times n$ diagonal matrix with the $w_i$'s on the diagonal ($W_{1,1}=w_1, W_{2,2}=w_2$ etc. and all other entries are zero). The prediction is then given by $\hat{y}=\theta^T \hat{x}$. If we later want to make a prediction about another point, $x'$, we need to recalculate the set of weights and obtain a new set of $\theta$'s.

Create a function which implements LWR by solving the system of equations given above (see Jupyter notebook `vika_01.ipynb` for details on how to do this with NumPy). Assume a local model on the form $f_\theta(x)=\theta_0 + \theta_1 x$.

Use LWR to fit the data in the `global.csv` dataset from problem 1), after scaling the $x$ values into $[0,1]$ with $\tilde{x}^{(i)}=x^{(i)}/\max_{k=1 \ldots n}[x^{(k)}]$ to avoid numerical issues. Create a scatter plot which shows the original data points and superimpose predictions for each $x$ value using $\sigma=0.02$ (this value is not optimal for this data set, more on that later).

*Comments*: The following functions are useful: `numpy.dot`, `numpy.linalg.norm`, `numpy.diag`, `math.exp`. Note that the `norm` and `diag` functions require NumPy arrays as inputs (rather than lists).

```python
import math
import numpy as np

def lwr_predict(X,y,xpred,sigma):
    # Locally weighted regression
    #    X ....... Input variables
    #    y ....... Output variable
    #    xpred ... Point to predict
    #    sigma ... Weight parameter (positive)
    # Return value
    #    ypred ... Predicted value at xpred
    assert(sigma>0)
    n = len(X)
    xpred = (xpred-min(X))/(max(X)-min(X)) #Skölum xpred á [0,1] líkt og X
    X = (X-min(X))/(max(X)-min(X)) #Skölum X gögnin á [0,1]
    X = np.c_[np.ones(n),X] # bætum við ásum í X-fylkið til að fá mat á theta_0
    W = np.zeros((n,n))
    for i in range(0,n):
        W[i,i]=np.exp(-(np.linalg.norm(xpred-X[i,1])**2)/(2*sigma**2))
    theta = np.linalg.solve(X.T @ W @ X ,X.T @ W @y)
    # Insert code here
    # ...
    ypred = theta[0] + theta[1]*xpred
 
    return ypred

X,y = np.loadtxt('global.csv', delimiter=',', unpack=True,skiprows=1)

sigma=0.02
ypred = np.zeros(len(X))
for i in range(0,len(X)):
    ypred[i] = lwr_predict(X,y,X[i],sigma)

plt.scatter(X,y)
plt.plot(X,ypred,c='r')

# Load data, predict and plot
# Insert code here
# ...
```

3) [Model selection, 25 points] This is a continuation of the two previous problems. Here you will use ''repeated subsampling'' to determine an appropriate value of a hyper-parameter (we use this strategy here rather than $k$-fold cross-validation since it is easier to implement). The method is as follows:
    
1. Iterate over a range of hyper-parameter values.<br>
2. $\quad$ Repeat 100 times:
3. $\qquad$ Create a permutation of the dataset
4. $\qquad$ Set 20% of the (permuted) points aside to use as a test set.
5. $\qquad$ Use the remaining data points to fit a model using the method of choice (LWR, k-NN reg. etc.)
6. $\qquad$ Evaluate the MSE on the test set
7. $\quad$ Calculate the average MSE for the 100 trials and store in a vector
8. $\quad$ (this becomes the prediction error estimate for the current value of the hyper-parameter)
9. Return the hyper-parameter value that corresponds to the smallest MSE

Implement the above model selection procedure for a given list of hyper-parameter values (see outline below).

For both $k$-NN regression and LWR perform the following: Create a figure which shows how the MSE varies with the value of the hyper-parameter ($k$ or $\sigma$). *Comments*: If you get the error message `Singular Matrix` when solving $(X^T W X)\theta = (X^T W)y$ for $\theta$, the value of $\sigma$ may be too small. For LWR, make sure that you are using the scaled $X$-values.

The `math.floor` and `math.ceil` functions are handy for computing indices into the train and validation sets. The function `numpy.average` returns the average of an array. To create a numpy array of evenly spaced values use `numpy.linspace`. To locate the smallest element in an array use `numpy.argmin`. To create a randomly permuted copy of the dataset $(X,y)$ you can use

```python
import numpy as np
def model_selection(func, X, y, testfr, parvals):
    # model selection based on random subsampling
    #    func .... Predictor function
    #    X ....... Input variables
    #    y ....... Output variable
    #    testfr .. Fraction of data to use as test set (0 < testfr < 1)
    #    parvals . Values of hyper-parameter
    # Return values
    #    mse ..... Vector of MSE values corresponding to different hyper-parameter values
    #    bestpar . Best value of the hyper-parameter
    assert(0 < testfr and testfr < 1)
    assert(func=='lwr' or func=='knn')
    n=len(X)
    mse = np.zeros(len(parvals))
    k = len(parvals)
    for i in range(0,k):    
        for j in range(0,100):
            perm=np.random.permutation(len(y))
            Xperm=X[perm]
            yperm=y[perm]
            m = int(testfr*n)
            X_test=Xperm[:m]
            y_test=yperm[:m]
            X_train=Xperm[m:]
            y_train=yperm[m:]
            ypred = np.zeros(m)
            err = np.zeros(100)
            if func=='knn':
                for s in range(0,m):
                    ypred[s] = knn_reg_predict(X_train, y_train, X_test[s], parvals[i])
            if func=='lwr':
                for s in range(0,m):
                    ypred[s] = lwr_predict(X_train,y_train,X_test[s],parvals[i])
            err[j] = np.mean((ypred-y_test)**2)
        mse[i] = np.mean(err)
    bestpar = parvals[np.where(mse==min(mse))]
  
    return (mse, bestpar)

# Load data, perform model selection, predict and plot
# Insert code here
X,y = np.loadtxt('global.csv', delimiter=',', unpack=True,skiprows=1)
n=len(X)
parvals_lwr = np.linspace(0.004,0.08,20)
mse_lwr, bestpar_lwr = model_selection('lwr',X,y,0.2,parvals_lwr)
parvals_knn = np.arange(2,30,1)
mse_knn, bestpar_knn = model_selection('knn',X,y,0.2,parvals_knn)
print('bestpar fyrir lwr',bestpar_lwr)
print('mse fyrir lwr',mse_lwr)
print('bestpar fyrir knn',bestpar_knn)
print('mse fyrir knn', mse_knn)

plt.plot(parvals_lwr,mse_lwr,'g')
plt.title('lwr-MSE vs. sigma')
plt.xlabel('sigma')
plt.ylabel('MSE')
plt.show(True)
plt.plot(parvals_knn,mse_knn,'r')
plt.xlabel('K')
plt.ylabel('MSE')
plt.title('knn-MSE vs. K')
plt.show(True)
ypred_knn = np.zeros(n)
ypred_lwr = np.zeros(n)
for i in range(0,n):
    ypred_knn[i] = knn_reg_predict(X,y,X[i],int(bestpar_knn))
    ypred_lwr[i] = lwr_predict(X,y,X[i],bestpar_lwr)
plt.plot(X,ypred_knn,'b')
plt.plot(X,ypred_lwr,'r')
plt.title('Samanburður á ályktunargildum fyrir knn og lwr')
plt.xlabel('X')
plt.ylabel('y-ályktun')
    
```

Use the following variables as inputs: `season yr mnth holiday weekday workingday weathersit atemp hum windspeed`. The output variable is the total number of bikes rented each day (variable `cnt`). The $R^2$ coefficient is a normalized measure of model fit that is often used in regression. A value of 1.0 indicates perfect fit whereas values close to zero or negative values indicate a bad fit. The coefficient is defined as $R^2=1-$MSE_model_fit / MSE_total where MSE_model_fit is the mean square error of the model with respect to the (test) data and MSE_total is the variance in the $y$-values. This value is provided by the regression object's `score` function.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

data = np.loadtxt('bike_sharing.csv', delimiter=',', unpack=False,skiprows=1)
X = np.c_[data[:,1:8],data[:,9:12]]
y = data[:,14]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_test)

print('Coefficients: \n', regr.coef_)
print('MSE:', mean_squared_error(y_test,y_pred))
print('R squared:', r2_score(y_test,y_pred))

plt.scatter(y_test,y_pred)

```
